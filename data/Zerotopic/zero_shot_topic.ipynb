```python
import pandas as pd
from konlpy.tag import Okt,Kkma
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import csv
from pororo import Pororo
from kss import split_sentences

zsl = Pororo(task="zero-topic", lang="ko")

filename='zero_shot_topic_221203.csv'
f = open(filename,'w',encoding='utf-8-sig', newline='')
writer = csv.writer(f)

df = pd.read_csv('clustered_article_221203_600.csv',names=['cluster_num','category','title','content','press','img','url'])

kkma = Kkma()


tf = ["긍정","부정"]
for j in range(1,df['cluster_num'].max()+1):
    temp_df = df[df['cluster_num']==j]
    for i, row in temp_df.iterrows():
        print(i,row['title'])
        sentence_arr = kkma.sentences(row['content'])
        t,f,n = 0,0,0
        for sent in sentence_arr:
            if len(sent)>512:
                sent_arr = sent.split('.')
                for s in sent_arr:
                    if len(s)>512:
                        continue
                    t += zsl(s,tf)[tf[0]]
                    f += zsl(s,tf)[tf[1]]
                n += len(sent_arr)
                continue
            t += zsl(sent,tf)[tf[0]]
            f += zsl(sent,tf)[tf[1]]
        result_t = t/(len(sentence_arr)+n-1)
        result_f = f/(len(sentence_arr)+n-1)
        writer.writerow([j,row['category'],row['title'],row['content'],row['press'],row['img'],row['url'],result_t,result_f])
        print("긍정:", result_t,"부정:", result_f)
        print()
f.close()

```
